# IMPORT LIBRARY

```python
from google_play_scraper import Sort, reviews
import pandas as pd
```

```python
# Ganti dengan ID aplikasi yang ingin kamu scrap, misalnya TikTok
app_id = 'com.ss.android.ugc.trill'

# Mengambil 100 ulasan terbaru
scrapreview, _ = reviews(
    app_id,
    lang='id',  # Bahasa Indonesia
    country='id',  # Negara Indonesia
    sort=Sort.NEWEST,  # Ulasan terbaru
    count=10000  # Jumlah ulasan yang ingin diambil
)
```

```python
# Membuat DataFrame dari hasil scraping
app_reviews_df = pd.DataFrame(scrapreview)

# Menyimpan ke file CSV lengkap
app_reviews_df.to_csv("ulasan_tiktok.csv", index=False)
```
